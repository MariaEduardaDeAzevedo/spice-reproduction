# Calculando correlações

## Pearson e Kendall-Tau

```python
metricas = ["bleu_4", "meteor", "spice", "spice-pt", "humano"]
```

```python
pearson_corr_pt = df_metricas_pt[metricas].corr(method="pearson")
pearson_corr_pt = pearson_corr_pt.reset_index().melt(id_vars="index")
pearson_corr_pt.columns = ["x", "y", "correlation"]

kendall_corr_pt = df_metricas_pt[metricas].corr(method="kendall")
kendall_corr_pt = kendall_corr_pt.reset_index().melt(id_vars="index")
kendall_corr_pt.columns = ["x", "y", "correlation"]


pearson_corr_eng = df_metricas_eng[metricas].corr(method="pearson")
pearson_corr_eng = pearson_corr_eng.reset_index().melt(id_vars="index")
pearson_corr_eng.columns = ["x", "y", "correlation"]

kendall_corr_eng = df_metricas_eng[metricas].corr(method="kendall")
kendall_corr_eng = kendall_corr_eng.reset_index().melt(id_vars="index")
kendall_corr_eng.columns = ["x", "y", "correlation"]
```

```python
def plot_heatmap(data, title):
    base = (
        alt.Chart(data)
        .mark_rect()
        .encode(
            x=alt.X("x:N", title="Variável"),
            y=alt.Y("y:N", title="Variável"),
            color=alt.Color(
                "correlation:Q", scale=alt.Scale(scheme="redblue", domain=(-1, 1))
            ),
            tooltip=["x", "y", alt.Tooltip("correlation:Q", format=".2f")],
        )
        .properties(width=300, height=300, title=title)
    )
    text = base.mark_text(baseline="middle", fontSize=12).encode(
        text=alt.Text("correlation:Q", format=".2f"),
        color=alt.condition(
            alt.datum.correlation > 0.5, alt.value("white"), alt.value("black")
        ),
    )
    return base + text
```

```python
heatmap_pearson_pt = plot_heatmap(pearson_corr_pt, "Correlação de Pearson")
heatmap_kendall_pt = plot_heatmap(kendall_corr_pt, "Correlação de Kendall Tau")

heatmap_pearson_eng = plot_heatmap(pearson_corr_eng, "Correlação de Pearson")
heatmap_kendall_eng = plot_heatmap(kendall_corr_eng, "Correlação de Kendall Tau")
```

```python
corr_chart_pt = alt.hconcat(heatmap_pearson_pt, heatmap_kendall_pt).properties(title="Correlação para amostras em Português")
corr_chart_eng = alt.hconcat(heatmap_pearson_eng, heatmap_kendall_eng).properties(
    title="Correlação para amostras em Inglês"
)

chart_corr = alt.vconcat(corr_chart_pt, corr_chart_eng)
chart_corr
```

## Gráficos de dispersão com Regressão Linear

```python
def scatter_plot_metrica(metrica, df):
    chart = alt.Chart(df).mark_point(filled=True, size=100).encode(
        x=alt.X(metrica, title=metrica.upper(), scale=alt.Scale(domain=[0, 1])),
        y=alt.Y("humano", title="Avaliação Humana", scale=alt.Scale(domain=[0, 1])),
        color="modelo:N",
        tooltip=["image_id", "modelo", metrica, "humano"],
        text="icon:N"
    ) + alt.Chart(df).transform_regression(
        metrica, "humano", groupby=["modelo"]
    ).mark_line().encode(
        x=metrica, y="humano", color="modelo:N"
    )
    return chart.properties(
        width=300, height=300, title=f"{metrica.upper()} vs. Avaliação Humana"
    )
```

```python
charts_a_pt = alt.hconcat(
    *(scatter_plot_metrica(m, df_metricas_pt) for m in ["bleu_4", "meteor"])
)
charts_b_pt = alt.hconcat(*(scatter_plot_metrica(m, df_metricas_pt) for m in ["spice", "spice-pt"]))
charts_pt = (
    alt.vconcat(charts_a_pt, charts_b_pt)
    .properties(title="Dispersão para amostras em Português")
)

charts_a_eng = alt.hconcat(
    *(scatter_plot_metrica(m, df_metricas_eng) for m in ["bleu_4", "meteor"])
)
charts_b_eng = alt.hconcat(
    *(scatter_plot_metrica(m, df_metricas_eng) for m in ["spice", "spice-pt"])
)
charts_eng = (
    alt.vconcat(charts_a_eng, charts_b_eng).properties(
        title="Dispersão para amostras em Inglês"
    )
)

chart_dispersao = (alt.vconcat(charts_pt, charts_eng).configure_axis(
        labelFontSize=15, titleFontSize=15, labelFont="Lato", labelColor="black"
    )
    .configure_legend(titleFontSize=15, labelFontSize=12)
    .configure_view(stroke=None))
chart_dispersao.display()
```

# MTC por métrica e modelo

## Média 

```python
medias_pt = (
    df_metricas_pt.groupby("modelo")[metricas]
    .mean()
    .reset_index()
)

medias_pt = medias_pt.melt(id_vars="modelo", var_name="métrica", value_name="valor")


medias_eng = df_metricas_eng.groupby("modelo")[metricas].mean().reset_index()

medias_eng = medias_eng.melt(id_vars="modelo", var_name="métrica", value_name="valor")
```

```python
bar_chart_media_pt = (
    alt.Chart(medias_pt)
    .mark_bar()
    .encode(
        x=alt.X(
            "métrica:N",
            title="Métrica",
        ),
        y=alt.Y("valor:Q", title="Média", scale=alt.Scale(domain=[0, 1])),
        color="modelo:N",
        column=alt.Column("modelo:N", title="Modelo"),
    )
    .properties(width=100, height=250, title="Médias para amostras em Português")
)

bar_chart_media_eng = (
    alt.Chart(medias_eng)
    .mark_bar()
    .encode(
        x=alt.X(
            "métrica:N",
            title="Métrica",
        ),
        y=alt.Y("valor:Q", title="Média", scale=alt.Scale(domain=[0, 1])),
        color="modelo:N",
        column=alt.Column("modelo:N", title="Modelo"),
    )
    .properties(width=100, height=250, title="Médias para amostras em Inglês")
)


bar_chart_media = (
    alt.hconcat(bar_chart_media_pt, bar_chart_media_eng)
    .properties(title="Médias")
)
bar_chart_media.display()
```

# Dispersão multivariada

```python
metricas_selecionadas_pt = df_metricas_pt[["modelo", "bleu_4", "meteor", "spice", "spice-pt", "humano", "icon"]]
metricas_selecionadas_eng = df_metricas_eng[
    ["modelo", "bleu_4", "meteor", "spice", "spice-pt", "humano", "icon"]
]
```

```python
base_pt = (
    alt.Chart(metricas_selecionadas_pt)
    .transform_fold(["bleu_4", "meteor", "spice", "spice-pt"], as_=["Métrica", "Valor"])
    .mark_point(filled=True, size=125)
    .encode(
        x=alt.X("Valor:Q", title="Valor da Métrica"),
        y=alt.Y("humano:Q", title="Avaliação Humana"),
        color="modelo:N",
        tooltip=["modelo:N", "Métrica:N", "Valor:Q", "humano:Q"],
    )
    .properties(width=250, height=250)
    .facet(facet=alt.Facet("Métrica:N", title="Métrica"), columns=2)
)

base_eng = (
    alt.Chart(metricas_selecionadas_eng)
    .transform_fold(["bleu_4", "meteor", "spice", "spice-pt"], as_=["Métrica", "Valor"])
    .mark_point(filled=True, size=125)
    .encode(
        x=alt.X("Valor:Q", title="Valor da Métrica"),
        y=alt.Y("humano:Q", title="Avaliação Humana"),
        color="modelo:N",
        tooltip=["modelo:N", "Métrica:N", "Valor:Q", "humano:Q"],
    )
    .properties(width=250, height=250)
    .facet(
        facet=alt.Facet(
            "Métrica:N",
            title="Métrica",
            header=alt.Header(labelFontSize=16, titleFontSize=18),
        ),
        columns=2,
    )
)

base = alt.hconcat(base_pt, base_eng)
base.display()
```
