# Calculando correlaÃ§Ãµes

## Pearson e Kendall-Tau

```python
metricas = ["bleu_4", "meteor", "spice", "spice-pt", "humano"]
```

```python
pearson_corr = df_metricas[metricas].corr(method="pearson")
pearson_corr = pearson_corr.reset_index().melt(id_vars="index")
pearson_corr.columns = ["x", "y", "correlation"]

kendall_corr = df_metricas[metricas].corr(method="kendall")
kendall_corr = kendall_corr.reset_index().melt(id_vars="index")
kendall_corr.columns = ["x", "y", "correlation"]
```

```python
def plot_heatmap(data, title):
    base = (
        alt.Chart(data)
        .mark_rect()
        .encode(
            x=alt.X("x:N", title="VariÃ¡vel"),
            y=alt.Y("y:N", title="VariÃ¡vel"),
            color=alt.Color(
                "correlation:Q", scale=alt.Scale(scheme="redblue", domain=(-1, 1))
            ),
            tooltip=["x", "y", alt.Tooltip("correlation:Q", format=".2f")],
        )
        .properties(width=300, height=300, title=title)
    )
    text = base.mark_text(baseline="middle", fontSize=12).encode(
        text=alt.Text("correlation:Q", format=".2f"),
        color=alt.condition(
            alt.datum.correlation > 0.5, alt.value("white"), alt.value("black")
        ),
    )
    return base + text
```

```python
heatmap_pearson = plot_heatmap(pearson_corr, "CorrelaÃ§Ã£o de Pearson")
heatmap_kendall = plot_heatmap(kendall_corr, "CorrelaÃ§Ã£o de Kendall Tau")
```

```python
corr_chart = alt.hconcat(heatmap_pearson, heatmap_kendall)
corr_chart.display()
```

## GrÃ¡ficos de dispersÃ£o com RegressÃ£o Linear

```python
def scatter_plot_metrica(metrica):
    chart = alt.Chart(df_metricas).mark_point(filled=True, size=80).encode(
        x=alt.X(metrica, title=metrica.upper(), scale=alt.Scale(domain=[0, 1])),
        y=alt.Y("humano", title="AvaliaÃ§Ã£o Humana", scale=alt.Scale(domain=[0, 1])),
        color="modelo:N",
        tooltip=["image_id", "modelo", metrica, "humano"],
        text="icon:N"
    ) + alt.Chart(df_metricas).transform_regression(
        metrica, "humano", groupby=["modelo"]
    ).mark_line().encode(
        x=metrica, y="humano", color="modelo:N"
    )
    return chart.properties(
        width=300, height=300, title=f"{metrica.upper()} vs. AvaliaÃ§Ã£o Humana"
    )
```

```python
charts_a = alt.hconcat(
    *(scatter_plot_metrica(m) for m in ["bleu_4", "meteor"])
)

charts_b = alt.hconcat(*(scatter_plot_metrica(m) for m in ["spice", "spice-pt"]))

charts = alt.vconcat(charts_a, charts_b)

charts.display()
```

# MTC por mÃ©trica e modelo

## MÃ©dia 

```python
medias = (
    df_metricas.groupby("modelo")[metricas]
    .mean()
    .reset_index()
)

medias = medias.melt(id_vars="modelo", var_name="mÃ©trica", value_name="valor")
```

```python
bar_chart_media = (
    alt.Chart(medias)
    .mark_bar()
    .encode(
        x=alt.X(
            "mÃ©trica:N",
            title="MÃ©trica",
        ),
        y=alt.Y("valor:Q", title="MÃ©dia", scale=alt.Scale(domain=[0, 1])),
        color="modelo:N",
        column=alt.Column("modelo:N", title="Modelo"),
    )
    .properties(width=100, height=250)
)

bar_chart_media.display()
```

## Mediana

```python
medianas = (
    df_metricas.groupby("modelo")[metricas]
    .median()
    .reset_index()
)

medianas = medianas.melt(id_vars="modelo", var_name="mÃ©trica", value_name="valor")
```

```python
bar_chart_mediana = (
    alt.Chart(medianas)
    .mark_bar()
    .encode(
        x=alt.X(
            "mÃ©trica:N",
            title="MÃ©trica"
        ),
        y=alt.Y("valor:Q", title="Mediana", scale=alt.Scale(domain=[0, 1])),
        color="modelo:N",
        column=alt.Column("modelo:N", title="Modelo"),
    )
    .properties(width=100, height=250)
)

bar_chart_mediana.display()
```

```python
chart_mtc = alt.hconcat(bar_chart_media, bar_chart_mediana)
chart_mtc
```

# DispersÃ£o multivariada

```python
metricas_selecionadas = df_metricas[["modelo", "bleu_4", "meteor", "spice", "spice-pt", "humano", "icon"]]
```

```python
base = (
    alt.Chart(metricas_selecionadas)
    .transform_fold(["bleu_4", "meteor", "spice", "spice-pt"], as_=["MÃ©trica", "Valor"])
    .mark_point(filled=True, size=125)
    .encode(
        x=alt.X("Valor:Q", title="Valor da MÃ©trica"),
        y=alt.Y("humano:Q", title="AvaliaÃ§Ã£o Humana"),
        color="modelo:N",
        tooltip=["modelo:N", "MÃ©trica:N", "Valor:Q", "humano:Q"],
    )
    .properties(width=250, height=250)
    .facet(facet=alt.Facet("MÃ©trica:N", title="MÃ©trica"), columns=2)
)

base.display()
```
